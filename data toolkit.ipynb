```python
1 # Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],   # Numeric column
    'Salary': [50000, 60000, 75000, 90000]
}

df = pd.DataFrame(data)

# Compute the mean of a specific column (e.g., 'Age')
mean_age = df['Age'].mean()

# Compute the mean of another column (e.g., 'Salary')
mean_salary = df['Salary'].mean()

# Print results
print(f"Mean Age: {mean_age}")
print(f"Mean Salary: {mean_salary}")
```

```python
4 #How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('your_dataset_csv')

# Compute the correlation matrix
corr_matrix = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap
sns.heatmap(
    corr_matrix,
    mask=mask,
    cmap=cmap,
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"shrink": 0.5}
)

# Add a title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()

```

```python
8 # Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Prepare data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create the plot
plt.plot(x, y1, label='Sine', color='blue', linestyle='-')
plt.plot(x, y2, label='Cosine', color='green', linestyle='--')
plt.plot(x, y3, label='Tangent', color='red', linestyle='-.')

# Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
9  #Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

# Create the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 32, 23, 45],
    'Salary': [50000, 60000, 55000, 70000]
}
df = pd.DataFrame(data)

# Define the threshold
salary_threshold = 55000

# Filter the DataFrame
filtered_df = df[df['Salary'] > salary_threshold]

# Display the filtered DataFrame
print(filtered_df)
```
